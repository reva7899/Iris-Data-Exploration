```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Loading dataset
iris=sns.load_dataset("iris")
print("iris")
iris.head()
iris.plot(kind='scatter', x="petal_length", y="petal_width")
plt.show()
iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
plt.show()
```

```python
#Visualization of the feature matrix(coloumn1-4) as an image
iris_modified=iris.iloc[:,0:4]
sns.heatmap(iris_modified)
            
```

```python
#2D scatter plots of the four features (creates a matrix of scatter plots for each pair of the feature)
sns.pairplot(iris, hue='species')
```

```python
#Plot of three features: sepal length, sepal width, petal width
fig=plt.figure()
scatter_plot=fig.gca(projection="3d")
scatter_plot.scatter(iris_modified['sepal_length'],iris_modified['sepal_width'],iris_modified['petal_width'])
scatter_plot.set_xlabel('sepal_length')
scatter_plot.set_ylabel('sepal_width')
scatter_plot.set_zlabel('petal_width')
features=iris.columns
sns.set_style('darkgrid')
sns.pairplot(iris[features])
plt.show()
```

```python
#Histogram for sepal length
plt.figure(figsize=(10,7))
x=iris["sepal_length"]
plt.hist(x, bins=20, color='blue')
plt.title('sepal_length in cm')
plt.xlabel('sepal_length')
plt.ylabel('count')
```

```python
#Histogram for sepal width
plt.figure(figsize=(10,7))
x=iris["sepal_width"]
plt.hist(x, bins=20, color='orange')
plt.title('sepal_width in cm')
plt.xlabel('sepal_width')
plt.ylabel('count')
```

```python
#Histogram for petal length
plt.figure(figsize=(10,7))
x=iris["petal_length"]
plt.hist(x, bins=20, color='green')
plt.title('petal_length in cm')
plt.xlabel('petal_length')
plt.ylabel('count')
```

```python
#Histogram for petal width
plt.figure(figsize=(10,7))
x=iris["petal_width"]
plt.hist(x, bins=20, color='brown')
plt.title('petal_width in cm')
plt.xlabel('petal_width')
plt.ylabel('count')
```

```python
#Boxplots for four features, parallel coordinate plot for four features 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
iris=sns.load_dataset("iris")
iris.head()
iris.describe()
iris=px.data.iris()
fig=px.parallel_coordinates(iris,color="species_id", labels={"species_id":"Species","sepal_width":"Sepal Width","sepal_length":"Sepal length","petal_width":"Petal Width","petal_length":"Petal Length"},color_continuous_scale=px.colors.diverging.Tealrose,color_continuous_midpoint=2)
nums_col=iris[['sepal_length','sepal_width','petal_length','petal_width']]
nums_col.boxplot()
iris.corr()
plt.figure(figsize=(6,6))
sns.heatmap(iris.corr())
fig.show()
```
